```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json('coursera_py.json')
df.head(10)
```

```python
skills = {}
for skillset in df['skills']:
    for skill in skillset:
        if skill not in skills.keys():
            skills[skill] = 1
        else:
            skills[skill] += 1
skills
```
